```python
import pandas as pd
import numpy as np
```

```python
exp_reward_board=[]
for agent, file in zip(exp_reward_history['AgentName'].values, exp_reward_history['ChoicesFile'].values):
    trials = pd.DataFrame.from_csv("/home/jupyter-pfeiffer/unbias/experiments/"+file)
    exp_reward_board.append((agent,np.sum(trials['AgentChoice'].values != trials['OutguessChoice'].values) ))
    

exp_reward_board = list(reversed(sorted(exp_reward_board, key=lambda x: x[1])))
```

```python
exp_choices_board=[]
for agent, file in zip(exp_choices_history['AgentName'].values, exp_choices_history['ChoicesFile'].values):
    trials = pd.DataFrame.from_csv("/home/jupyter-pfeiffer/unbias/experiments/"+file)
    exp_choices_board.append((agent,np.sum(trials['AgentChoice'].values != trials['OutguessChoice'].values) ))
    

exp_choices_board = list(reversed(sorted(exp_choices_board, key=lambda x: x[1])))
```
